## Understanding Missing `cabin_name` Values

Based on the unique values observed in `cabin_name` and `cabin_code_desc`, it appears that `cabin_name` is populated specifically when the `cabin_code_desc` is 'Economy'. The values in `cabin_name` ('Economy' and 'Economy Plus') likely specify sub-categories within the broader 'Economy' cabin class. ## Data Cleaning: Verifying Missing `cabin_name`

### Subtask:
Create a function to verify that all missing values in `cabin_name` correspond to non-'Economy' values in `cabin_code_desc`.

**Reasoning**:
Creating a function to verify this relationship will confirm our understanding of why `cabin_name` values are missing and ensure data consistency before proceeding with handling these missing values.

```python
def verify_missing_cabin_name(df):
    """
    Verifies that missing values in 'cabin_name' correspond to non-'Economy'
    values in 'cabin_code_desc'.

    Args:
        df (pd.DataFrame): The input DataFrame.

    Returns:
        bool: True if all missing 'cabin_name' values correspond to
              non-'Economy' in 'cabin_code_desc', False otherwise.
    """
    # Get rows where cabin_name is missing
    missing_cabin_name_rows = df[df['cabin_name'].isnull()]

    # Check if all of these rows have cabin_code_desc not equal to 'Economy'
    all_match = all(missing_cabin_name_rows['cabin_code_desc'] != 'Economy')

    return all_match

# Verify the relationship
if verify_missing_cabin_name(df):
    print("Verification successful: All missing 'cabin_name' values correspond to non-'Economy' in 'cabin_code_desc'.")
else:
    print("Verification failed: There are missing 'cabin_name' values that correspond to 'Economy' in 'cabin_code_desc'. Further investigation is needed.")
```

## Data Cleaning: Handling Missing `cabin_name` Values

### Subtask:
Replace missing values in the `cabin_name` column with 'Not Economy'.

**Reasoning**:
Replacing missing `cabin_name` values with 'Not Economy' is a suitable strategy for the ML model, as it treats the absence of a specific cabin name (outside of Economy) as a distinct and informative category.

```python
df['cabin_name'].fillna('Not Economy', inplace=True)

# Verify that there are no more missing values in 'cabin_name'
print("Missing values in 'cabin_name' after imputation:")
display(df['cabin_name'].isnull().sum())

# Display the value counts to see the new category
print("\nValue counts for 'cabin_name' after imputation:")
display(df['cabin_name'].value_counts())
```

## Data Cleaning: Imputing Missing `arrival_gate` using Route Information with Fallback

### Subtask:
Impute missing `arrival_gate` values based on the most common gate for the corresponding route (origin, destination, and departure gate), with a fallback to the most common gate for the destination station if the route is not found.

**Reasoning**:
This imputation strategy leverages the most specific information available (the full route) and falls back to a less specific but still relevant piece of information (the destination station) when the most specific information is not sufficient. This helps to fill in more missing values with potentially more accurate imputations.

```python
# Calculate the mode of 'arrival_gate' for each group of origin, destination, and departure gate
route_arrival_gate_mode = df.groupby(['origin_station_code', 'destination_station_code', 'departure_gate'])['arrival_gate'].transform(lambda x: x.mode()[0] if not x.mode().empty else None)

# Calculate the mode of 'arrival_gate' for each destination station as a fallback
destination_arrival_gate_mode = df.groupby('destination_station_code')['arrival_gate'].transform(lambda x: x.mode()[0] if not x.mode().empty else None)

# Calculate the mode of 'departure_gate' for each destination station as a second fallback
destination_departure_gate_mode = df.groupby('destination_station_code')['departure_gate'].transform(lambda x: x.mode()[0] if not x.mode().empty else None)

# Fill missing 'arrival_gate' values: first try the route mode, then fallback to destination arrival mode, then fallback to destination departure mode, finally use 'Unknown'
df['arrival_gate'].fillna(route_arrival_gate_mode, inplace=True)
df['arrival_gate'].fillna(destination_arrival_gate_mode, inplace=True)
df['arrival_gate'].fillna(destination_departure_gate_mode, inplace=True)


# Verify that there are fewer or no more missing values in 'arrival_gate'
print("Missing values in 'arrival_gate' after imputation using multiple fallbacks:")
display(df['arrival_gate'].isnull().sum())

# Check the value counts again to see the effect of imputation
print("\nValue counts for 'arrival_gate' after imputation using multiple fallbacks:")
display(df['arrival_gate'].value_counts(dropna=False))
```

# Cleaning Departure Gate

## Data Cleaning: Imputing Missing `departure_gate` using Route Information with Fallback

### Subtask:
Impute missing `departure_gate` values based on the most common gate for the corresponding route (origin, destination, and arrival gate), with fallbacks to the most common gate for the origin station and then the most common arrival gate for the origin station if the route is not found.

**Reasoning**:
Applying a similar multivariate imputation strategy to `departure_gate` as was used for `arrival_gate` leverages the relationships between origin/destination stations and gates to fill in missing values with potentially more accurate imputations based on observed patterns.

```python
# Calculate the mode of 'departure_gate' for each group of origin, destination, and arrival gate
route_departure_gate_mode = df.groupby(['origin_station_code', 'destination_station_code', 'arrival_gate'])['departure_gate'].transform(lambda x: x.mode()[0] if not x.mode().empty else None)

# Calculate the mode of 'departure_gate' for each origin station as a fallback
origin_departure_gate_mode = df.groupby('origin_station_code')['departure_gate'].transform(lambda x: x.mode()[0] if not x.mode().empty else None)

# Calculate the mode of 'arrival_gate' for each origin station as a second fallback (using arrival gate as a proxy if departure gate mode is not available for origin)
origin_arrival_gate_mode = df.groupby('origin_station_code')['arrival_gate'].transform(lambda x: x.mode()[0] if not x.mode().empty else None)


# Fill missing 'departure_gate' values: first try the route mode, then fallback to origin departure mode, then fallback to origin arrival mode, finally use 'Unknown'
df['departure_gate'].fillna(route_departure_gate_mode, inplace=True)
df['departure_gate'].fillna(origin_departure_gate_mode, inplace=True)
df['departure_gate'].fillna(origin_arrival_gate_mode, inplace=True)
df['departure_gate'].fillna('Unknown', inplace=True)


# Verify that there are fewer or no more missing values in 'departure_gate'
print("Missing values in 'departure_gate' after imputation using multiple fallbacks:")
display(df['departure_gate'].isnull().sum())

# Check the value counts again to see the effect of imputation
print("\nValue counts for 'departure_gate' after imputation using multiple fallbacks:")
display(df['departure_gate'].value_counts(dropna=False))
```
